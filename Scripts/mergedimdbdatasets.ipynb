```python
#merge IMDB Movies and IMDB data sets
imdb_movies = pd.read_csv("../Datasets/IMDb_movies.csv", dtype='unicode', encoding='utf8' )
imdb_ratings = pd.read_csv("../Datasets/IMDb_ratings.csv", dtype='unicode', encoding='utf8' )
merged_imdb_data = pd.merge(imdb_movies, imdb_ratings, on="imdb_title_id", how="inner")
merged_imdb_data.head()
```

```python
#convert merged data sets into CSV file
merged_imdb_data.to_csv("mergedimdb.csv")
```
