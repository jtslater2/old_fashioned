```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import data
df = pd.read_csv("../Final_Datasets/MergedTVandMovies.csv", dtype='unicode', encoding='utf8')
df.head()
```

```python
#Float Values
df["Year"] = df["Year"].astype(float)
df["IMDb"] = df["IMDb"].astype(float)
df["Netflix"] = df["Netflix"].astype(float)
df["Hulu"] = df["Hulu"].astype(float)
df["Prime Video"] = df["Prime Video"].astype(float)
df["Disney+"] = df["Disney+"].astype(float)
df
```

```python
df = df[["Title", "Year", "Netflix", "Hulu", "Prime Video", "Disney+"]]
df.describe()
```

```python
#Look at new shows 2018-2020 only
shows = df[(df["Year"] >= 2018)]
shows
```

```python
#Count of New Releases per Streaming Service
netflix_shows = shows.loc[shows["Netflix"] == 1]
hulu_shows = shows.loc[shows["Hulu"] == 1]
prime_shows = shows.loc[shows["Prime Video"] == 1]
disney_shows = shows.loc[shows["Disney+"] == 1]
print(len(netflix_shows))
print(len(hulu_shows))
print(len(prime_shows))
print(len(disney_shows))
```

```python
#Get Year and Service
netflix_shows = netflix_shows[['Year', 'Netflix']]
netflix_shows['Netflix'] = netflix_shows['Netflix'].replace([1], 'netflix')
netflix_shows.columns = ['Year', 'Service']


hulu_shows = hulu_shows[['Year', 'Hulu']]
hulu_shows['Hulu'] = hulu_shows['Hulu'].replace([1], 'hulu')
hulu_shows.columns = ['Year', 'Service']

prime_shows = prime_shows[['Year', 'Prime Video']]
prime_shows['Prime Video'] = prime_shows['Prime Video'].replace([1], 'prime')
prime_shows.columns = ['Year', 'Service']


disney_shows = disney_shows[['Year', 'Disney+']]
disney_shows ['Disney+'] = disney_shows ['Disney+'].replace([1], 'disney')
disney_shows.columns = ['Year', 'Service']
print(netflix_shows)
print(hulu_shows)
print(prime_shows)
print(disney_shows)
```

```python
#merge the source columns into a column
new_releases_df = pd.concat([netflix_shows, hulu_shows, prime_shows,  disney_shows], ignore_index=True, sort=False)

new_releases_df
```
